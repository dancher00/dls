(не забудьте передать scroing='roc_auc', чтобы при кроссвалидации сравнивались значения этой метрики, и refit=True, чтобы при потом модель обучилась на всем датасете с лучшим параметром C).

Потом pipeline надо передать в GridSearchCV. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
X = data[feature_cols].copy()
y = data[target_col].copy()


X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)

nan_after_encoding = X_encoded.isna().sum().sum()

if nan_after_encoding > 0:
    for col in X_encoded.columns:
        if X_encoded[col].isna().sum() > 0:
            X_encoded[col].fillna(X_encoded[col].median(), inplace=True)

scaler = StandardScaler()
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
```

```python

X_train, X_valid, y_train, y_valid = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

C_values = [100, 10, 1, 0.1, 0.01, 0.001]
best_score = 0
best_C = None

results_lr = []


for C in C_values:
    model = LogisticRegression(C=C, max_iter=1000, random_state=42, solver='lbfgs')
    model.fit(X_train, y_train)

    y_pred_proba = model.predict_proba(X_valid)[:, 1]
    score = roc_auc_score(y_valid, y_pred_proba)

    results_lr.append({'C': C, 'ROC-AUC': score})

    if score > best_score:
        best_score = score
        best_C = C

print(f"Лучший C: {best_C}")
print(f"Лучший ROC-AUC на validation: {best_score:.6f}")


lr_model = LogisticRegression(C=best_C, max_iter=1000, random_state=42, solver='lbfgs')
lr_model.fit(X_encoded, y)

y_train_pred_proba = lr_model.predict_proba(X_encoded)[:, 1]
train_score = roc_auc_score(y, y_train_pred_proba)
```

```python
# YOUR CODE

from catboost import CatBoostClassifier

X_cat = data[feature_cols].copy()
y_cat = data[target_col].copy()

print(f"Пропуски {X_cat.isna().sum().sum()}")
if X_cat.isna().sum().sum() > 0:
    for col in num_cols:
        if X_cat[col].isna().sum() > 0:
            X_cat[col].fillna(X_cat[col].median(), inplace=True)
    for col in cat_cols:
        if X_cat[col].isna().sum() > 0:
            X_cat[col].fillna(X_cat[col].mode()[0], inplace=True)

X_train_cat, X_valid_cat, y_train_cat, y_valid_cat = train_test_split(
    X_cat, y_cat, test_size=0.2, random_state=42, stratify=y_cat
)


cat_features_indices = [X_cat.columns.get_loc(col) for col in cat_cols]



base_model = CatBoostClassifier(
    iterations=100,
    learning_rate=0.1,
    random_state=42,
    verbose=False,
    eval_metric='AUC'
)

base_model.fit(
    X_train_cat, y_train_cat,
    cat_features=cat_features_indices,
    eval_set=(X_valid_cat, y_valid_cat),
    verbose=False
)

y_pred_base = base_model.predict_proba(X_valid_cat)[:, 1]
base_score = roc_auc_score(y_valid_cat, y_pred_base)

print(f"   Logistic Regression: {best_score:.6f}")
print(f"   CatBoost:  {base_score:.6f}")
```

```python
iterations_list = [50, 100, 200, 300]
learning_rates = [0.01, 0.05, 0.1, 0.2]

best_catboost_score = 0
best_params = {}

results_catboost = []


for n_iter in iterations_list:
    for lr in learning_rates:
        model = CatBoostClassifier(
            iterations=n_iter,
            learning_rate=lr,
            random_state=42,
            verbose=False,
            eval_metric='AUC'
        )

        model.fit(
            X_train_cat, y_train_cat,
            cat_features=cat_features_indices,
            eval_set=(X_valid_cat, y_valid_cat),
            verbose=False
        )

        y_pred = model.predict_proba(X_valid_cat)[:, 1]
        score = roc_auc_score(y_valid_cat, y_pred)

        results_catboost.append({
            'iterations': n_iter,
            'learning_rate': lr,
            'ROC-AUC': score
        })

        if score > best_catboost_score:
            best_catboost_score = score
            best_params = {'iterations': n_iter, 'learning_rate': lr}


print(f"Лучшие параметры: iterations={best_params['iterations']}, learning_rate={best_params['learning_rate']}")
print(f"Лучший ROC-AUC на validation: {best_catboost_score}")
```
